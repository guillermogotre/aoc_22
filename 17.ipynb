````python
import numpy as np

data = open('inputs/17.input').read().splitlines()
movement_list = list(map(lambda x: 1 if x == '>' else -1, data[0]))
shapes = open('inputs/17.shapes').read()
shapes_list = [np.array([[x == '#' for x in row] for row in e.splitlines()],dtype=bool)[::-1]
 for e in shapes.split('\n\n')]

class Loop:
    def __init__(self, data, return_index=False):
        self.data = data
        self.i = -1
        self.return_index = return_index
        
    def __next__(self):
        self.i = (self.i+1)%len(self.data)
        if self.return_index: return self.i, self.data[self.i]
        return self.data[self.i]
    
    def __iter__(self):
        return self
    
    def back(self, n=1):
        self.i -= n
        
    def forward(self, n=1):
        self.i += n
    
````
