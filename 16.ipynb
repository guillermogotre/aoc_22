````python
import re
from dataclasses import dataclass
from collections import deque
from functools import reduce
from queue import PriorityQueue

from typing import Any, Union

cache = {}
class Valve:
    def __init__(self,flow,name,dest):
        self.flow: int = flow
        self.name: str = name
        self.dest: dict = dest if isinstance(dest,dict) else {k:1 for k in dest}
        
    def flow_in_t(self,t):
        return self.flow * t

data = open('inputs/16.input').read().splitlines()
valves = []
for d in data:
    flow = int(re.findall(r'[0-9]+', d)[0])
    name, *dest = re.findall(r'[A-Z]{2}', d)
    valves.append(Valve(flow,name,dest))

new_items = []
for v in valves:
    if v.flow > 0:
        new_items.append(Valve(v.flow,f'{v.name}_',v.dest))
        v.flow = 0
        v.dest = [*v.dest,f'{v.name}_']

valves += new_items
       
import numpy as np 
m = np.zeros((len(valves),len(valves)),dtype=int)
for i,v1 in enumerate(valves):
    for j,v2 in enumerate(valves):
        m[i,j] = v2.name in v1.dest
        
max_path = 14
res = [m]
for i in range(max_path):
    res.append((res[-1]@m))
res = np.stack(res)

dist_map = (res>0).argmax(axis=0)+1
# dist_map2 = np.argmax(np.array([np.linalg.matrix_power(m,i) for i in range(1,14)])>0,axis=0)+1
indexes = [i for i,v in enumerate(valves) if v.name.endswith('_') or v.name == 'AA']
m_mask = np.isin(np.arange(len(dist_map)),indexes)

dist_map = dist_map[m_mask][:,m_mask]
# Prevent looping
dist_map = dist_map+np.eye(len(dist_map))*1e2
# Prevent from going back to AA
valves_names = [valves[i].name for i in indexes]
dist_map[:,valves_names.index('AA')] = 1e2
flow_list = [valves[i].flow for i in indexes]

optim_valves = []
for name,dist_row,flow in zip(valves_names,dist_map,flow_list):
    optim_valves.append(Valve(
        flow,
        name,
        dict(zip(valves_names,dist_row))
    ))
````

````python
import numpy as np
import pandas as pd
from dataclasses import dataclass, field
from itertools import product
from heapq import heappush, heappop
from collections import Counter
# @dataclass

def has_underscore_duplicate(l):
    c = Counter(l)
    return any(c[k]>1 for k in c if '_' in k)
valves_map_priv = {k.name:k for k in optim_valves}
class Step:
    def __init__(self,names,t,pending_t=None,opened_mask=None,explored_score=0, history=None):
        self.names = names
        self.t = t
        if pending_t is None:
            pending_t = np.zeros(len(valves_map_priv),dtype=int)
            
        if opened_mask is None:
            opened_mask = np.zeros(len(valves_map_priv),dtype=bool)
            opened_mask = opened_mask | [v.flow == 0 for v in valves_map_priv.values()]
        # done_names = [name for name,w in zip(names,pending_t) if w == 0]
        done_names = names
        self.opened_mask = opened_mask | [d in done_names for d in valves_map_priv]
        
        self.pending_t = pending_t
        to_add = sum((valves_map_priv[name].flow_in_t(t) if w==0 else 0 for w,name in zip(pending_t,names)))
        self.explored_score = explored_score + to_add
        
        self.history = history or []
        for w,name in zip(pending_t,names):
            if w == 0:
                self.history = self.history +[(name,t)]
        
        # dest = []
        # for name in names:
        #     drop_name = f'{name}_'
        #     drop = [drop_name] if drop_name in [k for b,k in zip(self.opened_mask, valves_map_priv) if b] else []
        #     dest.append(list(set(valves_map_priv[name].dest) - set(drop)))
        
        # self._dest = filter(lambda x: not has_underscore_duplicate(x), product(*dest))
        # self.dest = sorted(dest, key=lambda x: sum((not name.endswith('_') for name in x)), reverse=True)
        
        # if self.opened_mask.all(): self.t = 0
    
    def destinations(self):
        match = [('AA', 13),
            ('AA', 13),
            ('DD_', 11.0),
            ('JJ_', 10.0),
            ('HH_', 6.0),
            ('BB_', 6.0),]
        if self.history == match: 
            print("AA")
        dest = []
        for name,w in zip(self.names,self.pending_t):
            if w == 0:
                dest.append(list([
                    e for e,b 
                    in zip(
                        valves_map_priv[name].dest.items(),
                        self.opened_mask) 
                    if not b]))
            else:
                dest.append([(name,w)])
            dest[-1].append((name,self.t)) #TODO : check if this is correct
                
        for d in product(*dest):
            names, w = zip(*d)
            if len(set(names)) != len(names): continue
            
            # Filter to_drop
            
            min_pending = min(w)
            t = self.t - min_pending
            pending_t = [e-min_pending for e in w]
            
            if (t - max(pending_t)) < 0: continue
            
            
            next_step = Step(
            names = names,
            t = t,
            pending_t=pending_t,
            opened_mask = self.opened_mask,
            explored_score = self.explored_score,
            history = self.history
            )
            
            yield next_step
            
    # def go_to(self,dest):
    #     assert len(dest) == len(self.names)
        

class UpperBound:
    def __init__(self, valves_map):
        # m = np.zeros((len(valves_map),len(valves_map)),dtype=int)
        # for i,k in enumerate(valves_map):
        #     for j,l in enumerate(valves_map):
        #         m[i,j] = l in valves_map[k].dest
                
        # max_path = 14
        # res = [m]
        # for i in range(max_path):
        #     res.append((res[-1]@m))
        # res = np.stack(res)
        self.valves_map = valves_map
        # self.dist_map = (res>0).argmax(axis=0)+1
        self.dist_map = np.stack([
            [e for e  in v.dest.values()]
            for v in valves_map.values()
        ])
        
        self._node_idx = {v:i for i,v in enumerate(valves_map)}
        self.flows = np.array([v.flow for v in self.valves_map.values()])
        self.cache = {}
    
    def cache_key(self,node_name,t,visited_mask):
        return (node_name,t,*visited_mask)
        
    def node_idx(self,node_name):
        return self._node_idx[node_name]
    
    def upper_score(self,node_name,t,visited_mask):
        cache_key = self.cache_key(node_name,t,visited_mask)
        if cache_key in self.cache:
            return self.cache[cache_key]
        elif t == 0:
            self.cache[cache_key] = 0
            return 0
        
        idx = self._node_idx[node_name]
        
        not_visited = ~visited_mask
        n_not_visited = not_visited.sum()
        
        if n_not_visited == 0:
            self.cache[cache_key] = 0
            return 0
        
        # Ideal score
        if (not_visited.sum() == 0):
            print('ideal')
        
        min_dist_row = self.dist_map[idx][not_visited].min()
        # opt = self.flows[not_visited]*(t - dist_row)
        # opt_order = np.argsort(opt)[::-1]
        
        # Ideal distance
        min_dist = self.dist_map[not_visited][:,not_visited]+np.eye(n_not_visited)*1e3
        # triu_idx = np.triu_indices(n_not_visited, k = 1)
        # shortes_dist = sorted(min_dist[triu_idx])
        shortes_dist = [min_dist_row]+sorted(min_dist.min(axis=1))[:-1]
        shortes_dist = np.cumsum(shortes_dist)

        # subtract = self.flows[not_visited][opt_order]*shortes_dist
        res = np.clip(
            (t-shortes_dist)*sorted(self.flows[not_visited])[::-1]
            ,a_min=0,a_max=None).sum()
        
        self.cache[cache_key] = res
        return res
    
    def step_upper_score(self, step):
        # if len(step.names) < 1:
        #     print("WHAT")
        # print(len(indiv_scores),indiv_scores)
        try:
            opened_mask = step.opened_mask.copy()
            pending_name = [name for name,w in zip(step.names,step.pending_t) if w != 0]
            for name in pending_name:
                opened_mask[self.node_idx(name)] = False
            indiv_scores = [self.upper_score(name, step.t, opened_mask) for name in step.names]
            res = sum(indiv_scores)
            return res
        except:
            for name in step.names:
                self.upper_score(name, step.t, step.opened_mask)
            print(step)
            raise

from queue import PriorityQueue
def solve(valves_map, t, n=1):
    upper_bound = UpperBound(valves_map)
    # q = deque()
    # q = PriorityQueue()
    q = []
    counter = 0
    def put_in_q(step):
        nonlocal counter
        counter += 1
        # q.put((counter, step))
        heappush(q, (-step.explored_score,step.t,-upper_bound.step_upper_score(step),counter, step))
    def take_from_q():
        return heappop(q)[-1]
        
    start = Step(tuple(['AA']*n),t)
    # q.append(start)
    put_in_q(start)
    solutions = [start]
    
    while q:
        # step = q.popleft()
        step = take_from_q()
        # step = q.get()[-1]
        potential_score = step.explored_score + upper_bound.step_upper_score(step)
        best_explored = solutions[-1].explored_score
        
        if (step.explored_score > best_explored):
            solutions.append(step)
        if potential_score <= best_explored or step.t == 0:
            continue
        
        for next_step in step.destinations():
            if next_step.explored_score + upper_bound.step_upper_score(next_step) > best_explored:
                # q.append(next_step)
                put_in_q(next_step)
    return solutions
                    
import time

begin = time.time()
solution = solve(valves_map_priv,30,1)[-1]
print("Part 1", time.time()-begin, solution.explored_score)


begin = time.time()
solution = solve(valves_map_priv,26,2)[-1]
print("Part 2", time.time()-begin, solution.explored_score)

````
